```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
poverty = pd.read_csv('/Users/jAir/Desktop/Poverty_16.csv', index_col=1)
```

```python
poverty_sub = poverty[['Total Population','All Ages', 'Age 0-17',  
                                 'Age 5-17 in Families',
                                 'Age 0-4', 'Median Household Income', 'Welfare', 'DorR']]
poverty_sub['perc_pop_pov'] = poverty_sub['All Ages']/poverty_sub['Total Population']
poverty_sub = poverty_sub.iloc[1:51, :]
poverty_sub['welfare_dollars_per_person'] = (poverty_sub['Welfare']*1000000000)/poverty_sub['Total Population']
poverty_sub['z_score_perc_pov'] = (poverty_sub.perc_pop_pov - poverty_sub.perc_pop_pov.mean())/poverty_sub.perc_pop_pov.std(ddof=0)
poverty_sub['perc_pop_child'] = poverty_sub['Age 0-4']/poverty_sub['Total Population']
poverty_sub['z_score_perc_child'] = (poverty_sub.perc_pop_child - poverty_sub.perc_pop_child.mean())/poverty_sub.perc_pop_child.std(ddof=0)
```

```python
sns.lmplot(x = 'perc_pop_pov', y = 'welfare_dollars_per_person', data = poverty_sub)
plt.show()
```

```python
sns.lmplot(x = 'perc_pop_pov', y = 'welfare_dollars_per_person', data = poverty_sub, hue='DorR', palette='Set1')
plt.show()
```

```python
sns.lmplot(x = 'perc_pop_pov', y = 'welfare_dollars_per_person', data = poverty_sub, col='DorR')
plt.show()
```

```python
sns.residplot(x = 'perc_pop_pov', y = 'welfare_dollars_per_person', data = poverty_sub, color='indianred')
plt.show()
```

```python
sns.stripplot(y='perc_pop_pov', data=poverty_sub)
plt.ylabel('State Poverty Rates (%)')
plt.show()
```
